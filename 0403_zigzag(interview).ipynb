라이브 서비스를 운영중인 IT 스타트업에서 서비스의 현황을 빠르게 파악하는 것은 매우 중요합니다. 모든 구성원들이 서비스의 현황을 실시간으로 파악할 수 있어야 회사의 현 상태를 신속하게 점검할 수 있고, 문제를 발견하여 빠르게 수정할 수 있습니다.

가령 이번 달 목표 매출을 달성하지 못할 것 같으면, 마케팅팀과 기획팀에서는 목표 매출을 달성하기 위한 다양한 아이디어를 제안할 것이며, (기획전 주최, 할인 쿠폰 발행 등) 순간적으로 매출이나 페이지뷰가 급감할 경우 개발팀과 운영팀에서는 서비스가 기술적으로 문제가 없는지 점검해야 합니다. (서버 불안정, 특정 스마트폰 디바이스에서 주요 기능이 동작하지 않는 문제 등) 이를 빠르게 파악하지 못하면 회사의 의사결정 속도가 느려지고, 결국 회사가 목표를 달성하지 못하거나 경쟁사에게 추월당할 수 있습니다.

일반적으로 IT회사에서 대시보드를 구현하는 방법은 크게 두 가지로 나뉩니다. 첫 번째는 엑셀(내지는 구글 스프레드시트)을 활용해 구현하는 것이고, 두 번째는 프로그래밍 언어 + 웹 프레임워크를 활용해 자체 대시보드를 개발하는 것입니다. 두 가지 모두 장단점이 있습니다. 먼저 엑셀을 활용해 대시보드를 구현하는 방식의 장점은

1. 엑셀은 다른 툴에 비해 상대적으로 쉽기 때문에 모든 구성원들이 자유롭게 대시보드의 기능을 추가하거나 수정할 수 있습니다.
2. 엑셀에는 이미 구현된 기능이 많기 때문에, 새로운 기능을 추가하는데 소요되는 시간을 단축할 수 있습니다.

반면에 엑셀을 활용해 대시보드를 구현하는 방식의 단점은

1. 데이터의 양이 많아지면 대시보드가 현격하게 느려집니다. 회사마다 다르지만, 하루에 최소 1만 명 이상 방문하는 서비스의 경우 엑셀로 대시보드를 만들 경우, 엑셀 공식을 잘 튜닝하지 않으면 속도가 느려져서 사용하기가 매우 불편합니다.
2. 엑셀에서 지원하지 않는 기능이 필요할 경우, 이를 구현하는게 불가능하거나 구현하는데 너무 오랜 시간이 걸립니다.

두 번째 방식은 프로그래밍 언어 + 웹 프레임워크를 활용해 자체 대시보드를 개발하는 것입니다. 가령 파이썬의 경우, 쟝고([django](https://djangoproject.com))나 플라스크([Flask](http://flask.pocoo.org/)), SQL알케미([SQLAlchemy](https://www.sqlalchemy.org)), 그리고 자바스크립트 시각화 프레임워크인 [D3.js](https://d3js.org/) 등을 활용해 웹 대시보드를 자체 구현할 수 있습니다.

이렇게 대시보드를 자체 구현하는 방식의 장점은

1. 파이썬 프로그래밍 언어로 작성하였기 때문에 엑셀에 비해서 확연하게 속도가 빠릅니다. 파이썬으로 대시보드를 구현했을 때 속도가 느려서 곤란한 상황을 겪을 가능성은 거의 없습니다.
2. 다소 구현 난이도가 높은 기능도 파이썬과 관련 패키지를 활용하면 엑셀에 비해 어렵지 않게 구현할 수 있습니다. 또한 이미 운영중인 자사의 IT서비스와 대시보드를 연계하는 것도 굉장히 편리합니다.

반면에 프로그래밍 언어 + 웹 프레임워크를 활용해 대시보드를 자체 구축하는 방식의 단점은

1. 새로운 기능을 추가하는데 시간이 오래 걸립니다. 일단 프로그래밍 언어가 익숙하지 않은 직군(ex: 마케팅팀, 운영팀)의 경우에는 대시보드의 기능을 수정할 수가 없습니다.
2. 설령 마케팅, 운영팀이 기능을 대시보드의 기능을 수정했다고 하더라도, 이를 대시보드가 운영중인 서버에 배포하는 것은 매우 어렵습니다.
3. 대시보드를 구현한 프로그래밍 팀 입장에서도, 기능을 수정한 뒤 배포하는데 다소 시간이 걸리기 때문에 신속하게 기능을 추가하거나 변경하는 것이 어렵습니다.

다행히도 우리에게는 이 두 가지 방식의 장점을 모두 살릴 수 있는 아이디어가 있습니다. 바로 대시보드를 **쥬피터 노트북에 구현하는 것** 입니다. 이렇게 하면 쥬피터 노트북에서 실시간으로 기능을 구현해서 대시보드를 띄울 수 있습니다. 순서는 다음과 같습니다.

1. 쥬피터 노트북을 원격 서버에 띄웁니다. (원격 서버에 띄우는 방식은 [다음의 링크](https://jupyter-notebook.readthedocs.io/en/stable/public_server.html)를 참고해주세요)
1. 개발팀에서는 회사의 데이터를 쥬피터 노트북에서 접근할 있는 기본 세팅을 해줍니다. 가장 쉬운 방법은 DB 연결 권한을 준 뒤 쥬피터 노트북에서 바로 SQL 쿼리를 날리는 겁니다. ([read_sql](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_sql.html) 참조)
1. 개발팀에서 대시보드에 필요한 핵심 기능을 구현하되, 코멘트를 상세하게 달아서 비개발 직군 팀이 변수 몇 개만 바꾸면 바로 기능을 수정할 수 있도록 만들어줍니다.

이렇게 쥬피터 노트북에서 대시보드를 구현할 경우의 장점은 다음과 같습니다.

1. 파이썬과 판다스로 구현하였기 때문에 속도가 빠릅니다. 웹 대시보드를 직접 구현하는 것과 거의 동일한 실행 속도로 동작합니다.
2. 대시보드를 구현한 개발팀이 코멘트를 상세하게 남겨주면, 비개발 직군(ex: 마케팅, 운영)에서 간단한 기능을 추가하거나 수정하는 것은 큰 문제가 없습니다. (가령 11월 매출 현황을 시각화한 그래프를 활용하여 12월 매출 현황을 시각화하고 싶다면, 변수 몇 개만 바꿔주면 될 겁니다)
3. 복잡한 기능을 구현해야 할 경우, 개발팀이 쥬피터 노트북에서 바로 구현해주면 됩니다. 구현 결과를 대시보드 서버에 배포하는 등의 복잡한 행위를 할 필요가 없습니다.
